### Q1. What is AI Engineering, and how does it differ from ML Engineering?
**A:**
**ML Engineering:** Focuses on training and optimising ML models from data — feature engineering, model architecture, hyperparameter tuning, training infrastructure.

**AI Engineering:** Focuses on building production applications using pre-trained AI models (LLMs, diffusion models) — prompt engineering, RAG systems, agent architectures, LLMOps, evaluation.

**Key differences:**
| | ML Engineering | AI Engineering |
|--|---------------|----------------|
| Primary skill | Model training | Application building |
| Core artefact | Trained model | Prompts + retrieval + orchestration |
| Training | From scratch or fine-tuning | Usually just fine-tuning or none |
| Value creation | Better model performance | Better application behaviour |
| Tools | PyTorch, Scikit-learn | LangChain, LangGraph, CrewAI, vLLM |

**My answer:** I'm an AI Engineer — I build applications that use LLMs as reasoning engines (Nyaya-Pro, JobPilot AI, ExamGenie AI). I understand the underlying ML deeply enough to make good engineering decisions, but my value is in building reliable, production-grade AI applications.


### Q2. How do you decide whether a problem needs AI or a traditional software solution?
**A:** **Not everything needs AI.** Default to traditional software; only use AI when needed.

**Use AI when:**
- The problem requires understanding natural language at scale
- The rules are too complex/numerous to hand-code
- The input space is too variable (documents, images, free-text)
- The task requires generalisation to unseen inputs
- The cost of AI is justified by the business value

**Stick with traditional software when:**
- The problem has well-defined rules that can be coded explicitly
- You need 100% determinism and auditability
- Latency/cost constraints make LLM calls impractical
- The "AI" answer would just be if/else logic in disguise

**My framework:** "Can I write a unit test that this code should pass?" If yes, use traditional software. If the correct behaviour depends on nuanced language understanding or reasoning, use AI.

### Q4. How do you handle hallucinations when they occur in a production AI system?
**A:**
**Immediate (when a hallucination is reported):**
1. Log the specific case with full context (prompt, context, response)
2. Assess severity: is the hallucination causing real harm? If yes, disable the feature immediately.
3. Identify the root cause: wrong retrieval? Inadequate grounding instruction? Temperature too high?
4. Apply a targeted fix (tighter prompt, lower temperature, better retrieval)
5. Verify the fix on the golden test set before redeploying

**Systemic prevention:**
1. RAG with strict grounding instruction
2. Faithfulness checking pipeline (NLI or LLM-judge)
3. Mandatory citations in responses
4. Confidence calibration (require "I don't know" when uncertain)
5. Human review sampling (review 1% of outputs weekly)

### Q5. How do you decide between using an LLM API vs self-hosting an open-source model?

**Use LLM API (OpenAI, Gemini, Anthropic) when:**

- Small team — no one to manage GPU infrastructure

- High volume (cost savings kick in above ~1M tokens/day)

- Latency-critical and network round-trip is a problem
- Open-source model quality is sufficient for the task

### Q7. Describe your approach to debugging a poor-performing RAG system.
**A:** Systematic debugging — don't guess, measure.

**Step 1: Isolate the broken stage.**
- Retrieval OK, generation bad? → generation problem (prompt, LLM, faithfulness)
- Retrieval bad? → embedding, chunking, or hybrid search problem
- Both bad? → data quality or indexing problem

**Step 2: Examine retrieved chunks.**
- Print the top-3 retrieved chunks for failing queries. Are they relevant? If no → retrieval problem.
- If yes but answer still wrong → generation problem.

**Step 3: Diagnose retrieval.**
- Wrong chunks: try different embedding model, add BM25, fix chunk boundaries, add domain terms to queries
- Missing chunks: check if they're even in the index, check similarity threshold

**Step 4: Diagnose generation.**
- Hallucination: tighten system prompt, add grounding instruction, reduce temperature
- Wrong format: use native JSON mode + Pydantic validation
- Off-topic: check if retrieved context is too noisy (add reranking)

🏷️ *I debugged Nyaya-Pro's initial poor performance in exactly this order. The issue: chunks were too large (1024 tokens) → retrieval was imprecise. Fix: 512-token chunks + cross-encoder reranking → 40% accuracy improvement.*

### Q10. Tell me about a challenging AI project. What was the problem, approach, and outcome?
**A:** 🏷️ *"Nyaya-Pro — Agentic Legal RAG for Indian Law."*

**Problem:** Indian legal queries require understanding across multiple interconnected acts (BNS, BNSS, Constitution). Indian law is complex, recently changed (2023 BNS replacing IPC), and AI hallucinations are extremely costly in legal contexts.

**Approach:** 
1. Corpus ingestion — all 3 acts chunked at 512 tokens with 50-token overlap, embedded with `all-MiniLM-L6-v2`, stored in FAISS
2. Hybrid retrieval — BM25 for exact citations (e.g., "Section 302") + dense for semantic ("murder provisions")
3. Cross-encoder reranking — cut noise from top-10 to top-3 relevant chunks
4. Strict prompting — "Answer ONLY from context, cite every BNS/BNSS section you reference"
5. Pydantic output validation — every response validated against schema before returning

**Trade-offs:** Retrieval precision vs recall (chose higher precision via reranking), latency vs accuracy (reranking adds 200ms — worth it for legal accuracy).

**Outcome:** 94% faithfulness on golden test set (RAGAS), ~800ms median latency, near-zero hallucination rate in production.


### Q11. How would you handle a situation where an AI model produces biased or harmful outputs in production?
**A:**
**Immediate response:**
1. Assess severity — is ongoing harm happening? If yes, immediately disable the feature or add a kill switch.
2. Preserve evidence — log the specific harmful output, prompt, and context for post-mortem.
3. Notify stakeholders — don't hide it; transparent communication builds trust.
4. Implement temporary guardrail — even a blunt keyword filter to stop the immediate harm while a proper fix is developed.

**Root cause analysis:**
- Is it a training data bias?
- Is it a prompt that can be exploited to produce bias?
- Is it a retrieval problem (biased documents in the knowledge base)?

**Fix:**
- If prompt-induced: update the system prompt with explicit constraints
- If data-induced: fine-tune on debiased data; RLHF on preferred outputs
- If retrieval-induced: audit and clean the knowledge base

**Prevention going forward:**
- Add bias test cases to the golden test set
- Regular red-teaming
- Diverse evaluation raters


### Q12. How do you approach cost optimisation for an AI system that's exceeding budget?
**A:**
**Diagnose first:**
1. Which features/queries are the most expensive? (per-query cost breakdown)
2. What's the token breakdown? (input vs output, which prompts are longest?)
3. Are there cache opportunities being missed?

**Optimisation in order of impact:**
1. **Model downgrade** — test GPT-4o-mini or Gemini Flash for your use case (often 10-20× cheaper with 90% quality)
2. **Prompt caching** — long system prompts repeated across requests = huge savings
3. **Semantic caching** — cache responses for similar queries (20-40% cache hit rate)
4. **Shorten prompts** — audit prompt length; remove verbose instructions
5. **Reduce output tokens** — explicit length constraints, max_tokens cap
6. **Better retrieval** — retrieve fewer, more relevant chunks (reranking) → shorter context
7. **Model routing** — cheap model for simple queries, expensive for complex

**If still over budget:** reassess which features provide enough business value to justify the cost; consider removing low-value features.


### Q13. Describe a time when you had to choose between model accuracy and latency.
**A:** 🏷️ *"Nyaya-Pro's reranker trade-off."*

**Situation:** Initial RAG had no reranker — P50 latency: 400ms, faithfulness: 78%.

**Added cross-encoder reranker:** faithfulness jumped to 94%, but P50 latency went to 650ms. The legal domain required accuracy — a wrong legal citation can mislead someone in a court case.

**Decision:** The 250ms latency increase was worth the 16% accuracy improvement for a legal use case. In a consumer chatbot, I might have made the opposite call.

**Trade-off framework I use:**
- What's the cost of a wrong answer? (High in legal/medical → prioritise accuracy)
- What's the tolerance for latency? (Interactive chat: <1s target. Async processing: 10s fine.)
- Can I hide latency? (Streaming first tokens while reranking runs async)
- Is there a middle ground? (Lighter reranker, fewer candidates)

**Lesson:** Always define the latency budget AND the quality floor before designing. The design constraints change the entire solution.


### Q14. How would you handle a situation where your AI system's quality degrades over time?
**A:** "Model drift" in LLM applications can happen due to: underlying model updates by provider, shifting user query distribution, stale knowledge base, or accumulated prompt debt.

**Detection:**
- Continuous evaluation dashboard (faithfulness, user satisfaction, task completion rate)
- Automated alerts when metric drops > 5% from baseline
- Weekly manual spot-check of 20 sampled conversations

**Investigation checklist:**
1. Did the LLM provider update their model? (Check changelog, compare model version in logs)
2. Has the query distribution shifted? (New types of questions users are asking)
3. Is the knowledge base outdated? (Documents need refreshing)
4. Has the prompt aged poorly? (Newly relevant edge cases not covered)
5. Infrastructure change? (New server, different inference settings)

**Response:**
- If model update: evaluate new model vs old on golden test set; re-optimise prompt if needed
- If query shift: collect new examples, update golden test set, re-optimise for new patterns
- If stale KB: trigger re-indexing

### Q16. How would you approach building an AI feature with limited labeled data?
**A:** Limited labels ≠ can't build AI. Strategies:

1. **Few-shot prompting** — use powerful foundation model; 3-5 examples in the prompt. No training data needed.

2. **Synthetic data generation** — use GPT-4 to generate labelled examples for your specific task. 500 synthetic examples often sufficient for fine-tuning.

3. **Active learning** — build a small initial model → use uncertainty to select the most informative examples for human labelling → retrain → repeat. Get maximum value from each label.

4. **Zero-shot transfer** — use a model trained on a related task; evaluate on yours without any task-specific data.

5. **Weak supervision (Snorkel)** — define labelling functions (rules, heuristics) that are each imperfect; programmatically combine them to create noisy but scalable labels.

6. **Data flywheel** — ship v1 with few-shot prompting; log production inputs; have users implicitly label by accepting/rejecting outputs; use production data to train v2.

🏷️ *ExamGenie AI: started with zero labelled data. Used GPT-4 to generate (PDF excerpt → MCQ) pairs for synthetic training data. Then used those to fine-tune a cheaper model.*

### Q20. Your PM wants to ship an AI feature with a 15% hallucination rate on edge cases. How do you communicate the risk?

### Q22. You need to choose between a complex agentic system (15% better on benchmarks) vs a simpler RAG pipeline (easier to maintain). How do you decide?
**A:**
**My decision framework:**

**1. What's the actual user value of 15% improvement?**
- 15% better on benchmark ≠ 15% better for users. Benchmark gains often don't fully translate to real-world improvement.
- Run an A/B test or show both to a sample of users. Let data decide.

**2. What is the total cost of the complex system?**
- Engineering time to build and maintain
- Debugging complexity (agentic systems are much harder to debug)
- Failure modes (more complex = more ways to fail)
- Latency (multi-step agents are slower)

**3. What's the business context?**
- Startup with fast-moving requirements → choose simpler; you need to iterate fast
- Mature product with stable requirements and enough engineering bandwidth → complexity may be justified
- Safety-critical domain → simpler and more predictable is almost always better

**My default:** Ship the simpler RAG pipeline first. Measure real-world impact. Only add agent complexity when you have evidence the simpler approach's quality ceiling is a real business problem — not just a benchmark number.

*"Premature agentic complexity is the root of all AI system pain."*
