## ☁️ Top 30 AWS Questions for GenAI Engineers

### **1. What is AWS?**
**Answer:** AWS (Amazon Web Services) is a comprehensive, evolving cloud computing platform provided by Amazon that includes a mixture of infrastructure as a service (IaaS), platform as a service (PaaS) and packaged software as a service (SaaS) offerings.

### **2. Why use AWS instead of on-premise?**
**Answer:**
* **Scalability:** Instantly scale compute and storage resources up or down based on demand.
* **Pay-as-you-go:** Shift capital expenses (CapEx) to variable operational expenses (OpEx).
* **High Availability:** Global infrastructure with multiple Availability Zones (AZs) guarantees fault tolerance.
* **Managed Services:** Delegate operational tasks like patching, backups, and scaling to AWS.
* **Faster Deployments:** Provision complex architectures globally in minutes.

### **3. What is EC2?**
**Answer:** Elastic Compute Cloud (EC2) provides secure, resizable virtual servers (compute capacity) in the cloud.

**GenAI Example:** Hosting application servers (FastAPI/Flask), running local instances of lightweight models (via Ollama), or self-hosting databases.

**Follow-up (FastAPI Deployment Pipeline):**
$$\text{GitHub} \rightarrow \text{Docker} \rightarrow \text{Amazon EC2} \rightarrow \text{Nginx (Reverse Proxy)} \rightarrow \text{FastAPI (Uvicorn)}$$

### **4. What is S3?**
**Answer:** Simple Storage Service (S3) is an object storage service offering industry-leading scalability, data availability, security, and performance.

**GenAI Use Case:** Storing raw documents (PDFs, TXT), image assets, user uploads, model artifacts, and vector chunk data.

### **5. Difference between EC2 and S3**
**Answer:**
| Feature | Amazon EC2 | Amazon S3 |
|---|---|---|
| **Primary Role** | Compute (Virtual Servers) | Storage (Object Store) |
| **State** | Stateful (requires disk storage like EBS) | Serverless/Stateless (independent object store) |
| **Pricing** | Hourly/per-second compute runtime | Storage volume (per GB-month) + API requests |

### **6. What is Lambda?**
**Answer:** A serverless event-driven compute service that lets you run code for virtually any type of application or backend service without provisioning or managing servers.

**GenAI Example (Ingestion Pipeline):**
$$\text{User uploads PDF to S3} \rightarrow \text{S3 Event} \rightarrow \text{AWS Lambda (Extracts text, creates embeddings)} \rightarrow \text{Saves to Vector DB}$$

### **7. What is API Gateway?**
**Answer:** A fully managed service that makes it easy for developers to create, publish, maintain, monitor, and secure APIs at any scale.

**GenAI Architecture Flow:**
$$\text{Client/User} \rightarrow \text{API Gateway} \rightarrow \text{AWS Lambda / EC2} \rightarrow \text{Amazon Bedrock (LLM)}$$

### **8. What is RDS?**
**Answer:** Relational Database Service (RDS) makes it easy to set up, operate, and scale a relational database (such as PostgreSQL, MySQL, SQL Server) in the cloud.

### **9. What would you store in RDS for a chatbot?**
**Answer:** Structured relational application metadata, including:
* User credentials and authentication profiles
* Session metadata (IDs, timestamps, active flags)
* Chat history logs (orderly message chains mapped to users)
* Subscription states and transaction history

### **10. What is DynamoDB?**
**Answer:** A fully managed, multi-region NoSQL database service that provides fast, single-digit millisecond latency performance at any scale.

**Follow-up (DynamoDB vs PostgreSQL):**
* **DynamoDB:** NoSQL key-value store. Highly scalable horizontally, ideal for storing unstructured/semi-structured user session states and rapid write/read chat loops.
* **PostgreSQL:** Relational. Best for complex queries, joins, relational structures, and similarity searches using extensions like `pgvector`.

### **11. What is VPC?**
**Answer:** Virtual Private Cloud (VPC) lets you provision a logically isolated section of the AWS Cloud where you can launch AWS resources in a virtual network that you define.

**Follow-up (Why is VPC important?):** Security. It allows you to place backend services (databases, training servers) in private subnets, completely isolating them from the public internet, and only allowing traffic through designated public-facing gateways or load balancers.

### **12. What is IAM?**
**Answer:** Identity and Access Management (IAM) helps you securely control access to AWS resources. You use IAM to control who is authenticated (signed in) and authorized (has permissions) to use resources.

**Example:** Attaching an IAM Role to an EC2 instance that allows it to query an S3 bucket without embedding access keys inside your code.

### **13. What is a Security Group?**
**Answer:** A virtual firewall for your EC2 instances to control inbound and outbound traffic.

**Example:** Allowing inbound TCP traffic on ports 80 (HTTP), 443 (HTTPS), and 22 (SSH) from a specific IP address, while blocking all other ingress traffic.

### **14. What is CloudWatch?**
**Answer:** A monitoring and management service built for developers, system operators, site reliability engineers, and IT managers to monitor AWS resources and applications in real-time.

**GenAI Example:** Tracking LLM latency, monitoring API error rates, logging Bedrock token consumption, and setting up alarms for CPU utilization on GPU instances.

### **15. What is Auto Scaling?**
**Answer:** Auto Scaling monitors your applications and automatically adjusts capacity to maintain steady, predictable performance at the lowest possible cost.

**GenAI Example:** Automatically launching more compute instances when active user requests spike during a product launch, and tearing them down when traffic subsides.

### **16. What is Load Balancer?**
**Answer:** Elastic Load Balancing (ELB) automatically distributes incoming application traffic across multiple targets, such as Amazon EC2 instances, containers, IP addresses, and Lambda functions.

**Example:**
$$\text{Users} \rightarrow \text{Application Load Balancer (ALB)} \rightarrow \begin{cases} \text{FastAPI Server 1} \\ \text{FastAPI Server 2} \\ \text{FastAPI Server 3} \end{cases}$$

### **17. What is ECS?**
**Answer:** Elastic Container Service (ECS) is a fully managed container orchestration service that makes it easy to deploy, manage, and scale containerized applications on AWS.

### **18. What is EKS?**
**Answer:** Elastic Kubernetes Service (EKS) is a managed Kubernetes service that makes it easy to run Kubernetes on AWS without needing to install and operate your own Kubernetes control plane.

**Follow-up (ECS vs EKS):**
* **ECS:** Simple, highly integrated with the AWS ecosystem, lower management overhead. Best for standard web applications and API servers.
* **EKS:** Standards-based Kubernetes. More complex, offers deep customization, and is ideal for hybrid-cloud deployments or existing Kubernetes architectures.

### **19. What is Docker?**
**Answer:** An open-source platform that enables developers to package applications into containers—standardized executable components combining application source code with the operating system libraries and dependencies required to run it in any environment.

**Follow-up (Why Docker for GenAI?):** Consistency. GenAI apps often require complex local configurations (CUDA, PyTorch version alignments, specific OS packages). Docker guarantees that the exact environment tested locally runs identically on ECS or EKS without environment drift.

### **20. What is AWS Bedrock? (VERY IMPORTANT)**
**Answer:** A fully managed service that offers a choice of high-performing foundation models (FMs) from leading AI companies (like Anthropic, Meta, Mistral AI, Cohere, and Amazon) via a single API, along with a broad set of capabilities to build generative AI applications with security, privacy, and responsible AI.

**Follow-up (Why Bedrock instead of OpenAI?):**
* **Data Security:** Your data remains within your VPC; it is never shared with third-party model providers or used to train public base models.
* **AWS Integration:** Native IAM security controls, AWS CloudTrail logging for compliance, and unified billing.
* **Resilience:** Leverage AWS global scaling and region failovers directly.

### **21. What is SageMaker?**
**Answer:** A fully managed service that provides developers and data scientists with the ability to build, train, and deploy machine learning models quickly.

**Follow-up (SageMaker vs Bedrock):**
| Feature | Amazon SageMaker | Amazon Bedrock |
|---|---|---|
| **Primary Focus** | Custom ML model lifecycle (Build, train, tune, host) | Serverless Foundation Model access via API |
| **Infrastructure** | Provisioned instances/clusters (GPU config control) | Fully serverless (no instances to manage) |
| **Audience** | Data Scientists / ML Engineers | Application Developers / Software Engineers |

### **22. What is Vector Search?**
**Answer:** Searching for documents or data points by comparing dense numerical vector embeddings rather than exact keyword matching, relying on distance metrics like Cosine Similarity or L2 distance.

**AWS Service:** **Amazon OpenSearch Service** (with the Vector Search Engine plugin) or **Amazon Aurora** (using `pgvector`).

### **23. What AWS services would you use for RAG?**
**Answer:** An end-to-end Retrieval-Augmented Generation pipeline uses:
$$\text{S3 (Docs)} \rightarrow \text{Lambda (Chunking)} \rightarrow \text{Bedrock Titan (Embedding)} \rightarrow \text{OpenSearch/Aurora (Vector Store)} \rightarrow \text{Bedrock Claude (LLM Generation)}$$

### **24. Explain a GenAI Architecture on AWS**
**Answer:** A standard microservice setup:
$$\text{User} \rightarrow \text{API Gateway} \rightarrow \text{FastAPI on ECS Fargate} \rightarrow \text{Amazon Bedrock (LLM)} \leftrightarrow \text{Amazon OpenSearch (Vector DB)}$$

### **25. Where would you store PDFs?**
**Answer:** **Amazon S3** (Simple Storage Service) is the optimal storage layer due to its cost efficiency, durability, and integration with event triggers.

### **26. Where would you store embeddings?**
**Answer:**
* **Amazon OpenSearch Service** (dedicated Vector database engine)
* **Amazon Aurora PostgreSQL** (using `pgvector` extension)
* **SaaS Providers** (e.g., Pinecone, Weaviate, Milvus available via AWS Marketplace)

### **27. How would you secure an LLM API?**
**Answer:**
* Use **AWS IAM** policies to restrict Bedrock API invocations.
* Secure client requests through **API Gateway** using API keys, OAuth 2.0, or custom Lambda Authorizers (JWT validation).
* Keep communication traffic inside private networks by using **VPC Endpoints** (PrivateLink).

### **28. How would you scale a chatbot to 10,000 users?**
**Answer:**
* Deploy API backends on **ECS Fargate** configured with Auto Scaling rules.
* Position an **Application Load Balancer (ALB)** in front to distribute request loads.
* Use **Amazon ElastiCache for Redis** to cache repetitive answers and manage active session states.
* Request **Provisioned Throughput** in Amazon Bedrock to ensure dedicated inference capacity (Tokens Per Second) without throttling.

### **29. How would you reduce LLM cost?**
**Answer:**
* **Semantic Caching:** Store common questions/responses in Redis so you don't call the LLM for identical user requests.
* **Model Routing (Cascading):** Use smaller, cheaper models (e.g., Claude Haiku) for intent sorting or extraction, calling larger models (e.g., Claude Sonnet) only for complex reasoning.
* **Context Pruning:** Optimize prompts to send only the most relevant document chunks.

### **30. Design a Production GenAI System on AWS**
**Answer:** A fully fault-tolerant, scalable enterprise architecture:
$$\text{User} \rightarrow \text{CloudFront} \rightarrow \text{ALB} \rightarrow \text{FastAPI on ECS} \rightarrow \text{ElastiCache (Redis Cache)} \rightarrow \text{Bedrock (Claude)} \leftrightarrow \text{OpenSearch (Vector DB) / S3} \rightarrow \text{CloudWatch Monitoring}$$
