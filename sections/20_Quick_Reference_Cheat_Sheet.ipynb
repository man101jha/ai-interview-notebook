### One-Line Definitions for 50 Key Terms

| Term | One-line definition |
|------|---------------------|
| **Foundation model** | Giant model pre-trained on broad data; fine-tuned for specific tasks |
| **LLM** | Next-token prediction model trained on text at massive scale |
| **Transformer** | Neural network using self-attention to process sequences in parallel |
| **Tokenisation** | Splitting text into sub-word units (tokens) the model processes |
| **BPE** | Iteratively merge the most frequent character pairs into vocabulary tokens |
| **KV cache** | Stored past key-value pairs to avoid recomputing attention on each token |
| **Flash Attention** | I/O-aware attention that avoids materialising the N×N attention matrix |
| **RoPE** | Encodes position by rotating Q and K vectors; enables length extrapolation |
| **GQA** | Multiple query heads share fewer KV heads; reduces KV cache memory |
| **MoE** | Route each token to 1-2 expert FFNs; large parameters, low active compute |
| **Speculative decoding** | Draft small model → target model verifies in parallel; 2-3× speedup |
| **RAG** | Retrieve relevant documents → inject as context → LLM generates grounded answer |
| **FAISS** | Facebook's library for fast approximate nearest-neighbour vector search |
| **HNSW** | Hierarchical graph-based ANN index; O(log N) search time |
| **Hybrid search** | Dense vector search + sparse BM25 keyword search combined via RRF |
| **Reranker** | Cross-encoder that scores query+document pairs; more accurate than bi-encoder |
| **HyDE** | Embed a hypothetical ideal answer → use its embedding for retrieval |
| **GraphRAG** | Build a knowledge graph from docs; traverse entities for multi-hop answers |
| **Parent-child chunking** | Small chunks for retrieval precision; return larger parent chunks to LLM |
| **Paged Attention** | KV cache in non-contiguous pages; no over-allocation, supports sharing |
| **Continuous batching** | Add new requests to batch after each token; maximises GPU utilisation |
| **LoRA** | Freeze model; inject low-rank A×B matrices; train only those adapters |
| **QLoRA** | LoRA on 4-bit quantised base model; 7B fits in 6GB GPU |
| **RLHF** | SFT → reward model from human preferences → PPO optimisation |
| **DPO** | Align with preferences directly without a reward model or RL loop |
| **GRPO** | PPO variant using group mean as baseline; no separate critic needed |
| **Distillation** | Train small student to mimic large teacher's probability distributions |
| **ReAct** | Alternating Thought→Action→Observation loops for agentic reasoning |
| **MCP** | Open protocol for AI systems to use external tools standardly (like USB-C) |
| **Agent memory** | Short-term (context window) + long-term (vector DB) + episodic (logs) |
| **Prompt injection** | Malicious input overrides system instructions to hijack the AI |
| **Jailbreak** | Bypass safety training via creative prompting (roleplay, encoding, etc.) |
| **CoT** | Chain-of-thought: "Let's think step by step" improves multi-step reasoning |
| **Self-consistency** | Sample N CoT paths; majority vote on final answer; reduces errors |
| **Temperature** | Scales logits before softmax; higher = more random generation |
| **Top-p** | Sample from smallest token set summing to probability p; adaptive |
| **Perplexity** | exp(cross-entropy loss); lower = better language model |
| **BLEU** | N-gram precision between generated and reference text |
| **ROUGE** | N-gram recall between generated and reference text |
| **BERTScore** | Semantic similarity via BERT embeddings; captures synonyms |
| **RAGAS** | RAG evaluation framework: faithfulness, relevance, context precision/recall |
| **LLM-as-judge** | Use frontier LLM to score outputs; scales human-like evaluation |
| **Red teaming** | Systematically try to break your AI system before attackers do |
| **Differential privacy** | Mathematical guarantee that individual training records aren't leaked |
| **Hallucination** | Model generates confident but factually incorrect claims |
| **EU AI Act** | Risk-based AI regulation; high-risk systems need conformity assessment |
| **Quantisation** | Reduce weight precision (INT8, INT4) to shrink model and speed inference |
| **CLIP** | Trains text and image encoders contrastively; shared embedding space |
| **Diffusion model** | Learn to reverse a noising process; generate images by denoising |
| **LLMOps** | Operations for LLM applications: prompt versioning, monitoring, cost tracking |
