## 🧠 Machine Learning Basics (Q&A)

### **1. What is Machine Learning?**
**Answer:** A subset of AI that enables systems to learn patterns from data and make predictions without being explicitly programmed.

**Follow-up:** Difference between AI, ML, and Deep Learning?

### **2. Types of Machine Learning**
**Answer:**
* **Supervised Learning:** Training using labeled data (both input and output are known).
* **Unsupervised Learning:** Training using unlabeled data to discover hidden patterns.
* **Reinforcement Learning:** An agent learns by interacting with an environment and receiving rewards or penalties.

**Follow-up:** Give examples of each.

### **3. What is Supervised Learning?**
**Answer:** Training using labeled data where both input and output are known.

**Example:** House Features $\rightarrow$ House Price

### **4. What is Unsupervised Learning?**
**Answer:** Training using unlabeled data to discover hidden patterns.

**Example:** Customer Segmentation

### **5. What is Reinforcement Learning?**
**Answer:** An agent learns by interacting with an environment and receiving rewards or penalties.

**Example:** AlphaGo

### **6. What is Training Data?**
**Answer:** Data used to train a model.

### **7. What is Test Data?**
**Answer:** Data used to evaluate model performance on unseen examples.

### **8. Why split data into Train/Test?**
**Answer:** To measure generalization ability.

### **9. What is Overfitting?**
**Answer:** Model memorizes training data and performs poorly on new data.

**Example:** 98% Train Accuracy / 60% Test Accuracy

### **10. What is Underfitting?**
**Answer:** Model fails to learn important patterns.

### **11. How do you reduce Overfitting?**
**Answer:**
* More data
* Regularization
* Dropout
* Cross-validation
* Simpler model

### **12. What is Bias?**
**Answer:** Error due to incorrect assumptions.

### **13. What is Variance?**
**Answer:** Sensitivity to small data changes.

### **14. Explain Bias-Variance Tradeoff**
**Answer:** Good models balance bias and variance.

### **15. What is Cross Validation?**
**Answer:** Technique for evaluating model performance using multiple train-test splits.

**Follow-up:** What is K-Fold Cross Validation?

### **16. What is Feature Engineering?**
**Answer:** Creating useful features from raw data.

**Example:** Date $\rightarrow$ Day, Month, Weekday

### **17. What is Feature Scaling?**
**Answer:** Normalizing feature values into similar ranges.

### **18. Difference Between Normalization and Standardization**
**Answer:**
* **Normalization:** 0 to 1 range.
* **Standardization:** Mean = 0, Standard Deviation = 1.

### **19. What is Linear Regression?**
**Answer:** Predicts continuous values using a linear relationship.

**Example:** House Price Prediction

### **20. What is Logistic Regression?**
**Answer:** Used for classification problems.

**Example:** Spam vs Not Spam

### **21. Difference Between Regression and Classification**
**Answer:**
| Regression | Classification |
|---|---|
| Continuous output | Categorical output |
| House Price | Spam Detection |

### **22. What is a Decision Tree?**
**Answer:** Tree-like structure that splits data based on feature values.

### **23. What is Random Forest?**
**Answer:** Collection of multiple decision trees.

**Follow-up:** Why better than a single tree?

### **24. What is Clustering?**
**Answer:** Grouping similar data points.

**Example:** Customer Segmentation

### **25. What is K-Means?**
**Answer:** Popular clustering algorithm that groups data into K clusters.

### **26. What are Precision and Recall?**
**Answer:**
* **Precision:** Among predicted positives, how many are correct.
* **Recall:** Among actual positives, how many were found.

### **27. What is F1 Score?**
**Answer:** Harmonic mean of Precision and Recall.

### **28. What is Confusion Matrix?**
**Answer:** Shows:
* True Positive (TP)
* True Negative (TN)
* False Positive (FP)
* False Negative (FN)

### **29. What is Gradient Descent?**
**Answer:** Optimization algorithm used to minimize loss function.

**Follow-up:** How does it work? Move toward minimum error using gradients.

### **30. What is a Loss Function?**
**Answer:** Measures prediction error.

**Examples:** MSE, Cross Entropy
